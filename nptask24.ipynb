```python
import numpy as np
```

```python
file = np.genfromtxt(r"C:\Users\CHARISHMA\Downloads\Iris_clean_-_Copy.csv",dtype='float16',delimiter = ',')
file
```

```python
first_column = file[1:,0]
first_column
```

```python
np.nanmedian(first_column)
```

```python
missing_first = np.where(np.isnan(first_column),np.nanmedian(first_column),first_column)
missing_first
```

```python
cond = (missing_first<0) | (missing_first>8)
```

```python
sepal_length = np.where(cond,np.nanmedian(first_column),missing_first)
sepal_length
```

```python
second_column = file[1:,1]
second_column
```

```python
np.nanmedian(second_column)
```

```python
missing_second = np.where(np.isnan(second_column),np.nanmedian(second_column),second_column)
missing_second
```

```python
cond1 = (missing_second<0)|(missing_second>4)
```

```python
sepal_width = np.where(cond1,np.nanmedian(second_column),missing_second)
sepal_width
```

```python
third_column = file[1:,2]
third_column
```

```python
np.nanmedian(third_column)
```

```python
missing_third = np.where(np.isnan(third_column),np.nanmedian(third_column),third_column)
missing_third
```

```python
cond2 = (missing_third<0)|(missing_third>7)
```

```python
petal_length = np.where(cond2,np.nanmedian(third_column),missing_third)
petal_length
```

```python
fourth_column = file[1:,3]
fourth_column
```

```python
np.nanmedian(fourth_column)
```

```python
missing_fourth = np.where(np.isnan(fourth_column),np.nanmedian(fourth_column),fourth_column)
missing_fourth
```

```python
petal_width = np.where(cond3,np.nanmedian(fourth_column),missing_fourth)
petal_width
```
