##### array1

```python
np.mean(array1)  #mean
```

```python
array1.max()
```

```python
array1.min()
```

### spread

```python
#range
array1.max() - array1.min()
```

```python
#IQR
q1 = np.percentile(array1,25)
q1
```

```python
q3 = np.percentile(array1,75)
q3
```

```python
IQR = q3-q1
IQR
```

```python
lower = q1-(1.5*IQR)
lower
```

```python
upper = q3+(1.5*IQR)
upper
```

```python
#variance
np.var(array1)
```

```python
#standard deviation
np.std(array1)
```
